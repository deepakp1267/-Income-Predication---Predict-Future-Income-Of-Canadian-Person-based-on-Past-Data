```python
import numpy as np
import matplotlib.pyplot as plt

CanadaIncome = {
    "years": (1970,1971,1972,1973,1974,1975,1976,1977,1978,1979,1980,1981,1982,1983,1984,1985,1986,1987,1988,1989,1990,1991,1992,1993,1994,1995,1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016),
    "income": (3399.299037,3768.297935,4251.175484,4804.463248,5576.514583,5998.144346,7062.131392,7100.12617,7247.967035,7602.912681,8355.96812,9434.390652,9619.438377,10416.53659,10790.32872,11018.9558,11482.89153,12974.80662,15080.28345,16426.72548,16838.6732,17266.09769,16412.08309,15875.58673,15755.82027,16369.31725,16699.82668,17310.75775,16622.67187,17581.02414,18987.38241,18601.39724,19232.17556,22739.42628,25719.14715,29198.05569,32738.2629,36144.48122,37446.48609,32755.17682,38420.52289,42334.71121,42665.25597,42676.46837,41039.8936,35175.18898,34229.19363)
}

def gradient_descent(x, y):
    m_curr = b_curr = 0
    iterator = 10
    learning_rate = 0.05
    n = len(x)
    
    x_array = np.array(x)
    y_array = np.array(y)
    
    for i in range(iterator):
        y_pred = m_curr * x_array + b_curr
        
        
        cost = (1/2) * sum((y_array - y_pred) ** 2)
        md = -(1/n) * sum(x_array * (y_array - y_pred))
        bd = -(1/n) * sum(y_array - y_pred)
        
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        
        print("m_curr {}, b_curr {}, cost {}, iteration {}".format(m_curr, b_curr, cost, i))
    
    return m_curr, b_curr

m, b = gradient_descent(CanadaIncome["years"], CanadaIncome["income"])
y_pred = m * np.array(CanadaIncome["years"]) + b

plt.plot(CanadaIncome["years"], CanadaIncome["income"], color="green", label="Original Data")
plt.plot(CanadaIncome["years"], y_pred, color="blue", marker="*", label="Predicted Data")

plt.xlabel('Years')
plt.ylabel('Income')
plt.title('Linear Regression using Gradient Descent')
plt.legend()
plt.show()
```
